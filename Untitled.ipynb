```python
import numpy as np
import pandas as pd

#generete corr random varibles
from scipy.stats import multivariate_normal as mvn
#parameters
r=np.array([.2,.3,.4,.5])
mu=50
sd=15
corr={}
cov={}
chol={}
for i in range(len(r)):
    corr[i] = np.array([[1,r[i]],[r[i],1]])#corr matrix
    chol[i]= np.linalg.cholesky(corr[i]) 

np.random.seed(105)
rand_data = np.random.normal(mu, sd, size=(2,200))
corr_data_sc1 = pd.DataFrame(chol[0].dot(rand_data),index=["vis", "ver"]).T
corr_data_sc2 = pd.DataFrame(chol[1].dot(rand_data),index=["vis", "ver"]).T
corr_data_sc3 = pd.DataFrame(chol[2].dot(rand_data),index=["vis", "ver"]).T
corr_data_sc4 = pd.DataFrame(chol[3].dot(rand_data),index=["vis", "ver"]).T
print(corr_data_sc1.corr()) #check corr
print(corr_data_sc2.corr()) #check corr
print(corr_data_sc3.corr()) #check corr
print(corr_data_sc4.corr()) #check corr

print(corr_data_sc1.describe())
print(corr_data_sc2.describe())
print(corr_data_sc3.describe())
print(corr_data_sc4.describe()) 
```

```python
#add 10% noise
np.random.seed(15)
mu, sigma = 0, np.mean(corr_data_sc1)*0.1 
noise = np.random.normal(mu, sigma, corr_data_sc1.shape)
corr_data_sc1_10=(corr_data_sc1+noise)

mu, sigma = 0, np.mean(corr_data_sc2)*0.1 
noise = np.random.normal(mu, sigma, corr_data_sc2.shape)
corr_data_sc2_10=(corr_data_sc2+noise)

mu, sigma = 0, np.mean(corr_data_sc3)*0.1 
noise = np.random.normal(mu, sigma, corr_data_sc3.shape)
corr_data_sc3_10=(corr_data_sc3+noise)

mu, sigma = 0, np.mean(corr_data_sc4)*0.1 
noise = np.random.normal(mu, sigma, corr_data_sc4.shape)
corr_data_sc4_10=(corr_data_sc4+noise)
```

```python
#add 20% noise
np.random.seed(15)
mu, sigma = 50, np.mean(corr_data_sc1)*0.2 
noise = np.random.normal(mu, sigma, corr_data_sc1.shape)
corr_data_sc1_20=(corr_data_sc1+noise)

mu, sigma = 50, np.mean(corr_data_sc2)*0.2
noise = np.random.normal(mu, sigma, corr_data_sc2.shape)
corr_data_sc2_20=(corr_data_sc2+noise)

mu, sigma = 50, np.mean(corr_data_sc3)*0.2 
noise = np.random.normal(mu, sigma, corr_data_sc3.shape)
corr_data_sc3_20=(corr_data_sc3+noise)

mu, sigma = 50, np.mean(corr_data_sc4)*0.2 
noise = np.random.normal(mu, sigma, corr_data_sc4.shape)
corr_data_sc4_20=(corr_data_sc4+noise)
```

```python
#random and startigic allocator with 10% noise data
#split each scenorio into 2 sub groups for allocation step
sc1_r = corr_data_sc1.sample(frac = 0.5) #sc1_r will have random allocation
sc1_s = corr_data_sc1.drop(sc1_r.index) 
sc1_10_s = corr_data_sc1_10.drop(sc1_r.index) #strategic allocation 10% contamination

sc2_r = corr_data_sc2.sample(frac = 0.5) 
sc2_10_s = corr_data_sc2.drop(sc2_r.index) 
sc2_s = corr_data_sc2_10.drop(sc2_r.index) 

sc3_r = corr_data_sc3.sample(frac = 0.5)
sc3_s = corr_data_sc3.drop(sc3_r.index)
sc3_10_s = corr_data_sc3_10.drop(sc3_r.index)

sc4_r = corr_data_sc4.sample(frac = 0.5)
sc4_s = corr_data_sc4.drop(sc4_r.index)
sc4_10_s = corr_data_sc4_10.drop(sc4_r.index)

#simualting test for group with random allocators#
import random
n=int(len(sc1_10_r)/2) #number of pairs
rand_sc1 = {}#sc1 r=0.2
rand_sc2 = {}#sc2 r=0.3
rand_sc3 = {}#sc3 r=0.4
rand_sc4 = {}#sc4 r=0.5

for i in range(n):#loop simulation group selction and allocation
    g=sc1_r.sample(n = 2) #create random pairs
    x=round(random.random())
    p1=g.iloc[0, x]
    if x == 1: 
        y=0
    else:
        y=1
    p2=g.iloc[1, y]
    rand_sc1[i]=p1+p2

for i in range(n):
    g=sc2_r.sample(n = 2) #create random pairs
    x=round(random.random())
    p1=g.iloc[0, x]
    if x == 1: 
        y=0
    else:
        y=1
    p2=g.iloc[1, y]
    rand_sc2[i]=p1+p2

for i in range(n):
    g=sc3_r.sample(n = 2) #create random pairs
    x=round(random.random())
    p1=g.iloc[0, x]
    if x == 1: 
        y=0
    else:
        y=1
    p2=g.iloc[1, y]
    rand_sc3[i]=p1+p2

for i in range(n):
    g=sc4_10_r.sample(n = 2) #create random pairs
    x=round(random.random())
    p1=g.iloc[0, x]
    if x == 1: 
        y=0
    else:
        y=1
    p2=g.iloc[1, y]
    rand_sc4[i]=p1+p2


    #simualting test for group with s allocators#

n=int(len(sc1_s)/2) #number of pairs
st_sc1_10 = {}#sc1 r=0.2
st_sc2_10 = {}#sc2 r=0.3
st_sc3_10 = {}#sc3 r=0.4
st_sc4_10 = {}#sc4 r=0.5


count=0
count2=0
count3=0
count4=0

for i in range(n):#loop simulation group selction and allocation
    
    g=sc1_10_s.sample(n = 2) #create random pairs
    ind=g.index
    g1=sc1_s.loc[ind] ## g1 contains true score 
    
    if g['vis'].idxmax() == g['ver'].idxmax():#checking if same person
        count+=1
        #if same person use highest score of the two
        high=g.loc[g['vis'].idxmax()].idxmax()
        p1=g1[high][g[high].idxmax()]
        if high == 'vis':
            p2=g1['ver'][g['ver'].idxmin()]
        else:
            p2=g1['vis'][g['vis'].idxmin()]
    else:
        vis=g.loc[g['vis'].idxmax()].idxmax()
        p1=g1[vis][g[vis].idxmax()] #if diffrent participant, procced 
        ver=g.loc[g['vis'].idxmax()].idxmax()
        p2=g1[ver][g[ver].idxmax()]
    
    st_sc1_10[i]=p1+p2
    

for i in range(n):#loop simulation group selction and allocation
    
    g=sc2_10_s.sample(n = 2) #create random pairs
    ind=g.index
    g1=sc2_s.loc[ind] ## g1 contains true score 
    
    if g['vis'].idxmax() == g['ver'].idxmax():#checking if same person
        count2+=1
        #if same person use highest score of the two
        high=g.loc[g['vis'].idxmax()].idxmax()
        p1=g1[high][g[high].idxmax()]
        if high == 'vis':
            p2=g1['ver'][g['ver'].idxmin()]
        else:
            p2=g1['vis'][g['vis'].idxmin()]
    else:
        vis=g.loc[g['vis'].idxmax()].idxmax()
        p1=g1[vis][g[vis].idxmax()] #if diffrent participant, procced 
        ver=g.loc[g['vis'].idxmax()].idxmax()
        p2=g1[ver][g[ver].idxmax()]
    
    st_sc2_10[i]=p1+p2
    

for i in range(n):#loop simulation group selction and allocation
    g=sc3_10_s.sample(n = 2) #create random pairs
    ind=g.index
    g1=sc3_s.loc[ind] ## g1 contains true score 
    
    if g['vis'].idxmax() == g['ver'].idxmax():#checking if same person
        count3+=1
        #if same person use highest score of the two
        high=g.loc[g['vis'].idxmax()].idxmax()
        p1=g1[high][g[high].idxmax()]
        if high == 'vis':
            p2=g1['ver'][g['ver'].idxmin()]
        else:
            p2=g1['vis'][g['vis'].idxmin()]
    else:
        vis=g.loc[g['vis'].idxmax()].idxmax()
        p1=g1[vis][g[vis].idxmax()] #if diffrent participant, procced 
        ver=g.loc[g['vis'].idxmax()].idxmax()
        p2=g1[ver][g[ver].idxmax()]
    
    st_sc3_10[i]=p1+p2
    
for i in range(n):#loop simulation group selction and allocation
    
    g=sc4_10_s.sample(n = 2) #create random pairs
    ind=g.index
    g1=sc4_s.loc[ind] ## g1 contains true score 
    
    if g['vis'].idxmax() == g['ver'].idxmax():#checking if same person
        count4+=1
        #if same person use highest score of the two
        high=g.loc[g['vis'].idxmax()].idxmax()
        p1=g1[high][g[high].idxmax()]
        if high == 'vis':
            p2=g1['ver'][g['ver'].idxmin()]
        else:
            p2=g1['vis'][g['vis'].idxmin()]
    else:
        vis=g.loc[g['vis'].idxmax()].idxmax()
        p1=g1[vis][g[vis].idxmax()] #if diffrent participant, procced 
        ver=g.loc[g['vis'].idxmax()].idxmax()
        p2=g1[ver][g[ver].idxmax()]
    
    st_sc4_10[i]=p1+p2
     
   

print("Proportion of groups where single participant had both highest verbal and visual ability:")
print(f"when corr=0.2: {count/n}")
print(f"when corr=0.3: {count2/n}")
print(f"when corr=0.4: {count3/n}")
print(f"when corr=0.5: {count4/n}")

```
